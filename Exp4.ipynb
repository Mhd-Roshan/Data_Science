```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(3, 2))
plt.plot([1, 6], [3, 8], 'o')
plt.show()

plt.figure(figsize=(3, 2))
plt.plot([1, 6], [3, 8], 'o')
plt.show()

plt.figure(figsize=(3, 2))
plt.plot([3, 8, 1, 10], marker='o', ms=20, mec='black', mfc='yellow')
plt.show()

plt.figure(figsize=(3, 2))
plt.plot([80, 85, 90, 95, 100, 105, 110], [240, 250, 260, 270, 280, 290, 300])
plt.grid(color='green', linestyle='--', linewidth=0.5)
plt.show()

plt.figure(figsize=(3, 2))
plt.scatter(
    [5, 7, 8, 7, 2, 17, 2, 9, 4],
    [99, 86, 87, 88, 111, 86, 103, 87, 94],
    c=["red", "green", "blue", "yellow", "pink", "black", "orange", "purple", "beige"]
)
plt.show()

plt.figure(figsize=(3, 2))
plt.bar(["A", "B", "C", "D"], [4, 8, 3, 10], width=0.1, color='red')
plt.show()

plt.figure(figsize=(3, 2))
plt.pie([25, 35, 20, 20])
plt.show()
```
