```python
a=5
b=6
c=a+b
print("arithmetic operations")
print("sum is:",c)
d=a-b
print("difference is:",d)
e=a*b
print("multiplication:",e)
f=a/b
print("division:",f)

      
print("list operations")
l=[2,6,8,7,5]
print(l)
sum=0
for i in l:
    sum=sum+i
    print(i)
print("sum is:",sum) 
l.append(4)
print("list after inserted:",l)
l.remove(2)
print("list after removed:",l)
l.pop(2)
print("list after pop:",l)

      
print("set operations")
s1={"ami","rashi","ani"}
s2={"jaseem","aswin","shifana"}
print(s1)
print(s2)
s1.add("aysha")
print(s1)
s2.remove("aswin")
print(s2)

print("dictionary operations")
d={1:"apple",2:"orange",3:"grapes",4:"avocado"}
print(d)
print(d[1])
print(3 in d)

print("boolean operations")
t,f=True,False
print(t and f)
print(t or f)
print(not t)


print("String operations")
s="StrangerThings"
print("given string:",s)
print("length of given string:",len(s))
print("character in s[1]:",s[1])
print("to change all char to uppercase:",s.upper())
print("to change all char to lowercase:",s.lower())
```
